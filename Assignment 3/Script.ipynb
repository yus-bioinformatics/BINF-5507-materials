### k-mean clustering


```python
import numpy as np
import matplotlib.pyplot as plt
from sklearn.datasets import make_blobs, make_moons
from sklearn.cluster import KMeans
from sklearn.metrics import silhouette_score

# Define plotting function
def plot_clustering_results(X, labels, centers,silhouette, title):
    plt.figure(figsize=(8, 6))
    plt.scatter(X[:, 0], X[:, 1], c=labels, s=50, cmap='viridis')
    if centers is not None:
        plt.scatter(centers[:, 0], centers[:, 1], c='red', s=200, alpha=0.75, marker='X')
    plt.xlabel('Feature 1')
    plt.ylabel('Feature 2')
    plt.title(title)
    plt.grid(True)
    plt.show()


# Dataset 1: make_blobs 
X1, _ = make_blobs(n_samples=300, centers=3, cluster_std=[1.0, 2.5, 0.5], random_state=42)
#train K-Means on make_blobs dataset
kmeans1 = KMeans(n_clusters=4, max_iter=5, random_state=42)
kmeans1.fit(X1)
labels1 = kmeans1.predict(X1)
# Evaluate the model on make_blobs dataset
silhouette1 = silhouette_score(X1, labels1)
print(f"[make_blobs] Silhouette Score: {silhouette1:.3f}")
# Plot the clustering results for make_blobs
plot_clustering_results(X1, labels1, kmeans1.cluster_centers_, silhouette1, f"K-Means on make_blobs\nSilhouette Score: {silhouette1:.3f}")


# Dataset 2: make_moons (challenging for K-Means)
X2, _ = make_moons(n_samples=300, noise=0.05, random_state=42)
# Train K-Means on make_moons dataset
kmeans2 = KMeans(n_clusters=2, max_iter=5, random_state=42)
kmeans2.fit(X2)
labels2 = kmeans2.predict(X2)
# Evaluate the model on make_moons dataset
silhouette2 = silhouette_score(X2, labels2)
print(f"[make_moons] Silhouette Score: {silhouette2:.3f}")
# Plot the clustering results for make_moons
plot_clustering_results(X2, labels2, kmeans2.cluster_centers_, silhouette2, f"K-Means on make_moons\nSilhouette Score: {silhouette2:.3f}")
```

### Hierarchical Clustering

```python
from sklearn.cluster import AgglomerativeClustering
from scipy.cluster.hierarchy import dendrogram, linkage

# Dataset 1: make_blobs (for Hierarchical Clustering)
X1, _ = make_blobs(n_samples=300, centers=3, cluster_std=[1.0, 2.5, 0.5], random_state=42)

# Train an agglomerative clustering model
hierarchical1 = AgglomerativeClustering(n_clusters=3)
labels_h1 = hierarchical1.fit_predict(X1)
# Evaluate the model on make_blobs dataset
silhouette_h1 = silhouette_score(X1, labels_h1)
print(f"[make_blobs] Hierarchical Silhouette Score: {silhouette_h1:.3f}")

# Plot the clustering results for make_blobs
plot_clustering_results(X1, labels_h1, None, silhouette_h1, f"Hierarchical Clustering on make_blobs\nSilhouette Score: {silhouette_h1:.3f}")

# plot the dendogram 
Z = linkage(X1, method='ward')
plt.figure(figsize=(10, 7))
dendrogram(Z)
plt.title('Hierarchical Clustering Dendrogram (make_blobs)')
plt.xlabel('Sample Index')
plt.ylabel('Distance')
plt.show()

# Dataset 2: make_moons (challenging shape)
X2, _ = make_moons(n_samples=300, noise=0.05, random_state=42)
# Train an agglomerative clustering model
hierarchical2 = AgglomerativeClustering(n_clusters=2)
labels_h2 = hierarchical2.fit_predict(X2)
# Evaluate the model on make_moons dataset
silhouette_h2 = silhouette_score(X2, labels_h2)
print(f"[make_moons] Hierarchical Silhouette Score: {silhouette_h2:.3f}")
# Plot the clustering results for make_moons
plot_clustering_results(X2, labels_h2, None, silhouette_h2, f"Hierarchical Clustering on make_moons\nSilhouette Score: {silhouette_h2:.3f}")
    
```

### DBSCAN clustering 

```python
from sklearn.cluster import DBSCAN


# Dataset 1: make_blobs (for DBSCAN)
X1, _ = make_blobs(n_samples=300, centers=3, cluster_std=[1.0, 2.5, 0.5], random_state=42)
# Train a DBSCAN model  
dbscan1 = DBSCAN(eps=0.5, min_samples=5)        
labels_dbscan1 = dbscan1.fit_predict(X1)
# Evaluate the model on make_blobs dataset  
silhouette_dbscan1 = silhouette_score(X1, labels_dbscan1) if len(set(labels_dbscan1)) > 1 else -1
print(f"[make_blobs] DBSCAN Silhouette Score: {silhouette_dbscan1:.3f}")
# Plot the clustering results for make_blobs        
plot_clustering_results(X1, labels_dbscan1, None, silhouette_dbscan1, f"DBSCAN on make_blobs\nSilhouette Score: {silhouette_dbscan1:.3f}")  


# Dataset 2: make_moons (challenging shape for DBSCAN)
X2, _ = make_moons(n_samples=300, noise=0.05, random_state=42)  
# Train a DBSCAN model
dbscan2 = DBSCAN(eps=0.2, min_samples=5)    
labels_dbscan2 = dbscan2.fit_predict(X2)
# Evaluate the model on make_moons dataset
silhouette_dbscan2 = silhouette_score(X2, labels_dbscan2) if len(set(labels_dbscan2)) > 1 else -1
print(f"[make_moons] DBSCAN Silhouette Score: {silhouette_dbscan2:.3f}")
# Plot the clustering results for make_moons        
plot_clustering_results(X2, labels_dbscan2, None, silhouette_dbscan2, f"DBSCAN on make_moons\nSilhouette Score: {silhouette_dbscan2:.3f}")  
```
