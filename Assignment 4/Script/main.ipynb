1. ### Cox Proportional Hazards Regression


```python
#%pip install openpyxl

import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
from lifelines import CoxPHFitter

# load the dataset 
data = pd.read_excel('../Data/RADCURE_Clinical_v04_20241219.xlsx')


# Display the first few rows of the dataset
#print(data.head())


print(data['Status'].unique())

# Map event indicator from Status column
# map textual data to numeric values
data['Event'] = data['Status'].map({'Dead': 1, 'Alive': 0})  

# check the values of FU length
print(data['Length FU'].unique())
#check for missing values in Length FU
print(data['Length FU'].isnull().sum())

# check for  missing values in covariates
print(data[['Age','Tx Modality', 'Stage']].isnull().sum())
# Drop rows with missing values in key columns
df = data[['Length FU', 'Event','Age', 'Tx Modality', 'Stage']].dropna()

# Display the first few rows of the cleaned dataset
# print(df.head())
# df.shape


# Clean and normalize Tx Modality column
df['Tx Modality'] = df['Tx Modality'].astype(str).str.strip().str.replace(r'\s+', ' ', regex=True)

# Set 'RT alone' as the reference category
df['Tx Modality'] = pd.Categorical(
    df['Tx Modality'],
    categories=['RT alone', 'ChemoRT', 'RT + EGFRI', 'Postop RT alone'],
    ordered=True
)

# Convert categorical variables to dummy variables, dropping 'RT alone' and one stage
df_encoded = pd.get_dummies(df, columns=['Tx Modality', 'Stage'], drop_first=True)
#Display the first few rows of the encoded dataset
df_encoded.head()

# Fit the Cox model
cph = CoxPHFitter()
cph.fit(df_encoded, duration_col='Length FU', event_col='Event')

# View the results
cph.print_summary()

# Check proportional hazards assumption
cph.check_assumptions(df_encoded, p_value_threshold=0.05, show_plots=True)

```

### Random Survival Forests (RSF)

```python
from sksurv.ensemble import RandomSurvivalForest
from sklearn.inspection import permutation_importance
from sksurv.util import Surv


df_encoded = pd.get_dummies(df, columns=['Tx Modality', 'Stage'], drop_first=False)
X = df_encoded.drop(columns=['Length FU', 'Event'])
y = Surv.from_arrays(event=df['Event'].astype(bool), time=df['Length FU'])



# create random survival forest model
rsf = RandomSurvivalForest(n_estimators=50, random_state=42)  

# Fit the model
rsf.fit(X, y)   
# Print the model summary
print(rsf)

result = permutation_importance(rsf, X, y, n_repeats=15, random_state=42)

feature_importance = pd.DataFrame(
    {
        k: result[k]
        for k in ("importances_mean", "importances_std")
    },
    index=X.columns
).sort_values(by="importances_mean", ascending=False)

# Plot
plt.figure(figsize=(10, 6))
plt.title('Feature Importances (Permutation)')
plt.barh(feature_importance.index[:10][::-1],
         feature_importance['importances_mean'][:10][::-1],
         xerr=feature_importance['importances_std'][:10][::-1])
plt.xlabel('Mean Importance')
plt.ylabel('Features')
plt.tight_layout()
plt.grid(True)
plt.gca().invert_yaxis()
plt.show()



```

### concordance index

```python
from lifelines.utils import concordance_index


#Concordance index for Cox model
c_index_cox = concordance_index(df_encoded['Length FU'], 
                                -cph.predict_partial_hazard(df_encoded), 
                                df_encoded['Event'])
print(f"Cox model concordance index: {c_index_cox:.4f}")


# Concordance index for Random Survival Forest
c_index_rsf = concordance_index(df_encoded['Length FU'], -rsf.predict(X), df_encoded['Event'])
print(f"Random Survival Forest concordance index: {c_index_rsf:.4f}")   
```

### Kaplan-Meier Analysis 

###### # Comparing Treatment Type

```python
from lifelines import KaplanMeierFitter
from lifelines.statistics import logrank_test


# Filter the two treatment groups using the encoded DataFrame
group_rt = df[df["Tx Modality"] == "RT alone"]
group_chemo = df[df["Tx Modality"] == "ChemoRT"]

# Initialize Kaplan-Meier objects
kmf_rt = KaplanMeierFitter()
kmf_chemo = KaplanMeierFitter()

# Fit each group
kmf_rt.fit(durations=group_rt['Length FU'], event_observed=group_rt['Event'], label='RT Alone')
kmf_chemo.fit(durations=group_chemo['Length FU'], event_observed=group_chemo['Event'], label='ChemoRT')

# Plot survival curves
plt.figure(figsize=(10, 6))
kmf_rt.plot_survival_function()
kmf_chemo.plot_survival_function()
plt.title('Kaplan-Meier Survival Curves: RT Alone vs ChemoRT')
plt.xlabel('Time (years)')
plt.ylabel('Survival Probability')
plt.grid(True)
plt.legend()
plt.show()

# Perform log-rank test
results = logrank_test(group_rt['Length FU'], group_chemo['Length FU'],
                       event_observed_A=group_rt['Event'],
                       event_observed_B=group_chemo['Event'])

print(f"Log-rank test p-value: {results.p_value:.4f}")
```

###### # comparing Age groups 

```python
# Create age groups
group_young = df[df['Age'] <= 60]
group_old = df[df['Age'] > 60]

# Initialize KM fitters
kmf_young = KaplanMeierFitter()
kmf_old = KaplanMeierFitter()

# Fit each group
kmf_young.fit(group_young['Length FU'], event_observed=group_young['Event'], label='Age  60')
kmf_old.fit(group_old['Length FU'], event_observed=group_old['Event'], label='Age > 60')

# Plot
plt.figure(figsize=(10,6))
kmf_young.plot_survival_function()
kmf_old.plot_survival_function()
plt.title("Kaplan-Meier Survival by Age Group")
plt.xlabel("Time (years)")
plt.ylabel("Survival Probability")
plt.grid(True)
plt.legend()
plt.show()

# Log-rank test
results = logrank_test(
    group_young['Length FU'], group_old['Length FU'],
    event_observed_A=group_young['Event'],
    event_observed_B=group_old['Event']
)
print(f"Log-rank test p-value: {results.p_value:.4f}")
```
